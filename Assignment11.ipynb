```python
#Tugas No.1
import numpy as np

A = np.array([
    ["Denis","Scratch",8],
    ["Frank","Python",18],
    ["Roger","Unity",15]
]
)
print(A)
```

```python
#Tugas No.2
print("Baris Pertama =")
print(A[0])
print("------------------------")
print("Baris Kedua =")
print(A[1])
print("------------------------")
print("Baris Ketiga =")
print(A[2])
```

```python
#Tugas No.3
print(A[:,[1,2]])
```

```python
#Tugas No. 4

A[0,1] = "Roblox"
print(A)
```

```python
#Tugas No. 5.1

B = np.array ([
    ["Java","Richie",60,"C"],
    ["python","Neils",80,"A"],
    ["R","Brie",75,"B"]
])
print(B)
```

```python
#Tugas No. 5.2
print("Baris Pertama =")
print(B[0])
print("Baris Kedua =")
print(B[1])
print("Baris Ketiga =")
print(B[2])
```

```python
#Tugas No. 5.3
print("Kolom Pertama =")
print(B[:,0])
print("Kolom Kedua =")
print(B[:,1])
print("Kolom Ketiga =")
print(B[:,2])
```

```python
#Tugas No. 5.4
print(B[:2, 2:])


```

```python
#Tugas No. 5.5
import copy
C = copy.deepcopy(B)
C[:,0] = "Data Analyst"
print(C)
```

```python
#Tugas No. 5.6
D = copy.deepcopy(B)
D[2,2] = "90"
D[2,3] = "A+"
print(D)
```
