```python
#Tugas No. 2
score = 0
end_score = 10
while score < end_score:
  score += 1
  print("score anda adalah =", score)
if score == end_score:
  print("Selamat Anda Telah Memenangkan Permainan!")
```

```python
#Tugas No. 3
for x in range (5):
    for i in range (5):
      print("*", end = " ")
    print("\n")
for x in range (5):
    for i in range (x + 1):
      print("*", end = " ")
    print("\n")
```

```python
#Tugas No. 4
x = 0
i = int(input("masukkan sebuah bilangan = "))
hasil = 0
while x <= i:
  if x % 2 == 0:
    hasil += x
  x += 1

print("jumlah angka bilangan genap =",(hasil))
```

```python
#Tugas No. 5
Hasil = 0
awal = 0
batas = 10

while awal < batas:
  angka = int(input("masukkan angka = "))
  hasil += angka
  awal += 1

rata = hasil / batas
print("jumlah kesepuluh bilangan tersebut adalah =", (hasil))
print("rata rata adalah =", (rata))
```
