```python
import numpy as np
import pandas as pd
import io

url = ('https://raw.githubusercontent.com/gemara/data_analyst/main/bali_tourist.csv')
turis = pd.read_csv(url)
turis
```

```python
from pandas.core.groupby.groupby import T
#Tugas No. 1

d2018 = turis[turis['year'] == 2018]
d2019 = turis[turis['year'] == 2019]

bulan = ['January','February','March','April','May','June','July','August','September','October','November','December']

t2018 = d2018[bulan].sum()
t2019 = d2019[bulan].sum()

k2018 = t2018.sum()
k2019 = t2019.sum()

print(f"total jumlah turis 2018 = ", k2018)
print(f"total jumlah turis 2019 = ", k2019)
```

```python
#Tugas No. 2
mean_k2018 = k2018.mean()
mean_k2019 = k2019.mean()

print(f'rata-rata pengunjung tahung 2018 = ',round(mean_k2018))
print(f'rata-rata pengunjung tahung 2019 = ',round(mean_k2019))
```

```python
#Tugas No. 3

ma2018 = t2018.idxmax()
ma2019 = t2019.idxmax()
mi2018 = t2018.idxmin()
mi2019 = t2019.idxmin()

print(f'jumlah turis 2018 paling banyak pada bulan : ',ma2018)
print(f'jumlah turis 2019 paling banyak pada bulan : ',ma2019)
```

```python
#Tugas No. 4

print(f'jumlah turis 2018 paling sedikit pada bulan : ',mi2018)
print(f'jumlah turis 2019 paling sedikit pada bulan : ',mi2019)
```

```python
#Tugas No. 5

g2018 = d2018.groupby('area')[bulan].sum()

g2018.idxmax()

g2019 = d2019.groupby('area')[bulan].sum()

g2019.idxmax()
```

```python
#Tugas Tabel
tabel1 = d2018.groupby('year')[bulan].sum()
tabel2 = d2019.groupby('year')[bulan].sum()

tabel1['Max'] = ma2018
tabel1['Min'] = mi2018
tabel1['kedatangan setahun'] = round(k2018)
tabel1['rata-rata'] = round(mean_k2018)

tabel2['Max'] = ma2019
tabel2['Min'] = mi2019
tabel2['kedatangan setahun'] = round(k2019)
tabel2['rata-rata'] = round(mean_k2019)

final_tabel = pd.concat([tabel1, tabel2])

final_tabel
```
