```python
# Tugas No. 1
data_diri = {
    "Nama" : input(str("Masukkan Nama Kamu = ")),
    "Umur" : input(str("Umur Kamu Berapa? = ")),
    "Alamat" : input(str("Kamu Tinggal Di Mana? = ")),
    "Username" : input(str("Silahkan Pilih Kata Pengguna Kamu = ")),
    "Password" : input(str("Masukkan Password yang Kamu Pilih = "))
}
while True:
  pass_confirm = input(str("Ulangi Password yang kamu pilih = "))
  if pass_confirm == data_diri["Password"]:
      print("")
      print("password telah terkonfirmasi")
      print("")
      for x in data_diri:
        print(f"{x}, {data_diri[x]}")
      print("")
      break
  elif pass_confirm != data_diri["Password"]:
      print("")
      print("Password tidak sama, Silahkan Coba lagi cuy!")
      print("")

```

```python
# Tugas No. 2
koleksi = {
    "Fiksi" : "Forest Gump",
    "Non_Fiksi" : "Made in Japan",
    "Komik" : "City Hunter",
    "Pelajaran" : "Kalkulus"
}
print("Koleksi Buku Saat ini adalah : ")
print("")
for x in koleksi:
  print(f"{x}, {koleksi[x]}")
print("")
print("menu : ")
print("")
print("1. Menambahkan")
print("2. Menghapus")
print("3. Selesai")
print("")


while True:
  pilihan = int(input("Masukkan Pilihan = "))
  if pilihan == 1:
    jenis_buku = input("Masukkan Jenis Buku untuk ditambahkan = ")
    judul_buku = input("Masukkan Judul Buku untuk ditambahkan= ")
    koleksi.update({jenis_buku : judul_buku})
    print("")
    print(f"Anda telah menambahkan jenis buku {jenis_buku} dengan judul {judul_buku} ke dalam koleksi!")
    print("")
    print("Koleksi Buku Saat ini adalah : ")
    print("")
    for x in koleksi:
      print(f"{x}, {koleksi[x]}")
    print("")
    print("menu : ")
    print("1. Menambahkan")
    print("2. Menghapus")
    print("3. Selesai")
    print("")
  elif pilihan == 2:
    del_jenis_buku = input("Masukkan Jenis Buku untuk dihapus = ")
    del_judul_buku = input(f"Masukkan Judul Buku dari Jenis Buku {del_jenis_buku} untuk dihapus = ")
    if del_jenis_buku in koleksi:
      koleksi.pop(del_jenis_buku)
      print("")
      print(f"Anda telah menghapus buku dengan Jenis Buku {del_jenis_buku} yang berjudul {del_judul_buku} dari koleksi")
      print("")
    elif del_judul_buku not in koleksi:
      print("")
      print("Buku tidak ditemukan!, Silahkan Coba Lagi Ya!")
      print("")
    print("Koleksi Buku Saat ini adalah : ")
    for x in koleksi:
      print(f"{x}, {koleksi[x]}")
    print("")
    print("menu : ")
    print("1. Menambahkan")
    print("2. Menghapus")
    print("3. Selesai")
    print("")
  elif pilihan == 3:
    print("")
    print("")
    print("Selesai, Terima Kasih.")
    break
  elif pilihan != 1 or 2 or 3:
    print("")
    print("Nomor yang anda masukkan tidak terdaftar, Silahkan Ulangi!")
    print("")
```
