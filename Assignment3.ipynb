```python
#Tugas No. 1
nilai = int(input("Nilai hasil ujian adalah = "))

if(nilai < 0):
  print("silahkan masukan angka yang benar!")
else:
  print("kamu mendapatkan nilai ")
  if(nilai <= 50 and nilai > 0):
    print("D")
  if(nilai <= 70 and nilai > 50):
    print("C")
  if(nilai <= 80 and nilai > 70):
    print("B")
  if(nilai > 80):
    print("A")
```
