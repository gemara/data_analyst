```python
#Tugas No. 1
koleksi = {
    1 : "ayam",
    2 : "sapi",
    3 : "mawar",
    4 : "kambing",
    5 : "lily"
}

tumbuhan = ["mawar", "lily"]
hewan = ["ayam", "sapi", "kambing"]
key_hapus =[]
untuk_diproses = koleksi.copy()
print("sebelum proses, isi dictionary adalah :")
print(koleksi)
print("")
print("yang merupakan tumbuhan adalah :")
print(tumbuhan)

for key, value in koleksi.items():
  if value in tumbuhan:
    key_hapus.append(key)

for x in key_hapus:
  untuk_diproses.pop(x)

hasil_proses = untuk_diproses.copy()

print("")
print("setelah item tumbuhan dikeluarkan, isi dictionary adalah :")
print(hasil_proses)





```

```python
#Tugas No. 2
koleksi2 = {
    1 : "ayam",
    2 : "sapi",
    3 : "mawar",
    4 : "kambing",
    5 : "lily"
}
untuk_proses2 = koleksi2.copy()

for key, value in koleksi2.items():
  if key  % 2 == 1:
    untuk_proses2.update({key : "hewan"})

print("Sebelum proses, isi dictionary adalah :")
print(koleksi2)
print("")
print("setelah setiap key ganjil dalam dictionary diganti dengan kata hewan, maka isi dictionary adalah :")
print(untuk_proses2)


```
