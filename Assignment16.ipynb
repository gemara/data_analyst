```python
#Tugas No. 1
data_tugas = pd.Series(
    [408,943,902,117,389,435,793,356,651,993,786,157],
    index =["Januari","Februari","Maret","April","Mei","Juni",
            "Juli","Agustus","September","Oktober","November","Desember"]
    )
data_tugas
```

```python
#Tugas No. 2
data_tugas.mean()
```

```python
#Tugas No. 3 Langsung
rata_rata = data_tugas.mean()
data_tugas[data_tugas > rata_rata] = rata_rata
data_tugas
```

```python
#Tugas No. 3 Pakai For
for index,value in data_tugas.items():
  if value > rata_rata:
    data_tugas[index] = rata_rata

data_tugas
```
