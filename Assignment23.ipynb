```python
q1 = np.percentile(data['performance_score'], 25)
q2 = np.percentile(data['performance_score'], 50)
q3 = np.percentile(data['performance_score'], 75)

print(q1)
print(q2)
print(q3)

iqr = q3 - q1 #Interquartile Range

lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

print(lower_bound)
print(upper_bound)
```

```python
outlier = [x for x in data['performance_score'] if x < lower_bound or x > upper_bound]

print(outlier)
```

```python

condition1 = data['performance_score'] > lower_bound
condition2 = data['performance_score'] < upper_bound
data_bersih = data[(condition1 & condition2)]
data_bersih
```

```python
uplim = data['performance_score'].quantile(.99)
lowlim = data['performance_score'].quantile(.01)
data_bersih2 = data[(data['performance_score'] < uplim) & (data['performance_score'] > lowlim )]
data_bersih2
```
