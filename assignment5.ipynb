```python
# No. 6
mystr = "Welcome to Koding Akademi"
mystr2 = "Welcome to Koding Akademi"
print(mystr[11:])
print(mystr[:8])
print(mystr[::-1])

# Auto
pecah = mystr2.split()
masukan = input("masukkan kata = ")

for x in pecah:
  if masukan == x:
    panjang = len(x)
    awal = mystr2.index(masukan)
    print(mystr[awal : awal + panjang])
    print(f"slice {awal} : {awal + panjang}")
  else:
    print("Kata tidak ditemukan!")
    break

balik = pecah[::-1]
hasil = " ".join(balik)
print(hasil)
```
