```python
angka_proses = [1,2,3,4,5]
def tambah():
  try:
    num1 = int(input("Masukkan angka pertama = "))
    num2 = int(input("Masukkan angka kedua = "))
    hasil = (num1 + num2)
    print("")
    print("hasilnya adalah", hasil)
    print("")
  except ValueError:
    print("")
    print("Harus berupa angka!")
  except TypeError:
    print("")
    print("Harus berupa angka!")
  except ZeroDivisionError:
    print("")
    print("angka tidak boleh nol")
  finally:
    print("")
    print("proses selesai")

def kali():
  try:
    num1 = int(input("Masukkan angka pertama = "))
    num2 = int(input("Masukkan angka kedua = "))
    hasil = (num1 * num2)
    print("")
    print("hasilnya adalah", hasil)
    print("")
  except ValueError:
    print("")
    print("Harus berupa angka!")
  except TypeError:
    print("")
    print("Harus berupa angka!")
  except ZeroDivisionError:
    print("")
    print("angka tidak boleh nol")
  finally:
    print("")
    print("proses selesai")

def bagi():
  try:
    num1 = int(input("Masukkan angka pertama = "))
    num2 = int(input("Masukkan angka kedua = "))
    hasil = (num1 / num2)
    print("")
    print("hasilnya adalah", hasil)
    print("")
  except ValueError:
    print("")
    print("Harus berupa angka!")
  except TypeError:
    print("")
    print("Harus berupa angka!")
  except ZeroDivisionError:
    print("")
    print("angka tidak boleh nol")
  finally:
    print("")
    print("proses selesai")

def kurang():
  try:
    num1 = int(input("Masukkan angka pertama = "))
    num2 = int(input("Masukkan angka kedua = "))
    hasil = (num1 - num2)
    print("")
    print("hasilnya adalah", hasil)
    print("")
  except ValueError:
    print("")
    print("Harus berupa angka!")
  except TypeError:
    print("")
    print("Harus berupa angka!")
  except ZeroDivisionError:
    print("")
    print("angka tidak boleh nol")
  finally:
    print("")
    print("proses selesai")

while True:
  print("")
  print("Pilih proses yang akan dilakukan :")
  print("")
  print("1. Penjumlahan")
  print("2. Pengurangan")
  print("3. Perkalian")
  print("4. Pembagian")
  print("5. Keluar dari proses")
  print("")

  try:
    angka1 = int(input("Masukkan pilihan = "))
  except ValueError:
    print("")
    print("Error, Silahkan Coba Lagi!")
    print("")
  finally:


    if angka1 not in angka_proses:
      print("")
      print("Error, Silahkan Coba Lagi!")
      print("")
    elif angka1 in angka_proses:
      if angka1 == 1:
        tambah()
      elif angka1 == 2:
        kurang()
      elif angka1 == 3:
        kali()
      elif angka1 == 4:
        bagi()
      elif angka1 == 5:
        print("")
        print("Selamat Tinggal, Terima Kasih!")
        break





```
